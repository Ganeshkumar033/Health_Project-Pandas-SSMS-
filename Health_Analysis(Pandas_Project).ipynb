```python
#18. What is the median treatment cost by gender and diagnosis 
df.groupby(['Gender', 'Diagnosis'])['Treatment_Cost'].median()
```

```python
#20.What Percentage of patients are from “Cardiology”
(df['Department'].value_counts(normalize=True)['Cardiology']) * 100
```
